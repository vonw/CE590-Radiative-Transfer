# Malkmus Band Transmission Model

## This notebook approximates the transmission of the atmosphere using the equations for the Malkmus Random Band Model from Petty (2006), A First Course in Radiative Transfer.

```python
# y is the "grayness parameter" and is equal to aL / delta, where aL is the Lorentz line width 
#       and delta is the average line spacing.
ys  = array([1., 0.1, 0.01, 0.001, 0.0001])   # See Figure 10.3 in Petty (2006).
Sud = arange(0.1,1000,0.001)
```

```python
# Calculate the transmission using the Malkmus model.
for y in ys:
    T   = exp( (-pi*y/2.) * ( (1. + (4./(pi*y)) * Sud)**0.5 - 1. ) )
    semilogx(Sud,T)

xlabel('Su/d')
ylabel('Band Transmission')
title('Transmission from Malkmus Random Model')
legend(('1.0','0.1','0.01','0.001','0.0001'),loc='best')
```
