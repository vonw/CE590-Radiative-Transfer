# Reflection and Refraction

## Refraction

###Snell's Law

$$\sin(\theta_t) = (n_1/n_2)\sin(\theta_i)$$

####Case 1: $\theta_i = \theta_t = 0$ 

```python
N1 = 1.0   # air
N2 = 1.33  # water
ti = 0. * pi/180.
tt = arcsin((N1/N2) * sin(ti))
print 'The incident    angle is: ', ti * 180./pi
print 'The transmitted angle is: ', tt * 180./pi
```

####Case 2:   $\theta_i > 0;   N_2 > N_1$

```python
N1 = 1.0   # air
#N2 = 1.025  #Liquid He
#N2 = 1.33  # water
#N2 = 1.544  # Sodium Chloride
N2 = 2.419  # Diamond

ti = 30. * pi/180.
tt = arcsin((N1/N2) * sin(ti))
print 'The incident    angle is: ', ti * 180./pi
print 'The transmitted angle is: ', tt * 180./pi
```

####Case 3:   $\theta_i > 0;   N_2 < N_1$

```python
N1 = 1.33  # water
N2 = 1.0   # air
ti = 30. * pi/180.
tt = arcsin((N1/N2) * sin(ti))
print 'The incident    angle is: ', ti * 180./pi
print 'The transmitted angle is: ', tt * 180./pi
```

####Case 3:   $\theta_i > 0;   N_2 < N_1$   --  Special case -> Critical Angle

```python
N1 = 1.33  # water
N2 = 1.0   # air

# Critical angles
tc = arcsin(N2/N1)

tt = arcsin((N1/N2) * sin(tc))
print 'The incident    angle is: ', ti * 180./pi
print 'The transmitted angle is: ', tt * 180./pi

print 'The critical    angle is: ', tc * 180./pi
```

##Reflectivity

```python
# Rp is the reflectivity when the electric field vector is parallel to the plane of incidence.
# Rs is the reflectivity when the electric field vector is perpendicular to the plane of incidence.

N1 = 1.000  # air
#N2 = 1.025  #Liquid He
N2 = 1.33  # water
#N2 = 1.544  # Sodium Chloride
#N2 = 2.419  # Diamond
m = N2/N1
ti = arange(0.,91.,1.)
costt = ( 1 - (sin(ti*pi/180.)/2)**2 )**0.5
costi = cos(ti*pi/180.)

plot(ti,costi,ti,costt)
xlabel('Incident angle (degrees)')
ylabel('Cosine of incident and transmitted angles')
legend(('incident','transmitted'),loc='best')
```

```python
Rp = abs( (costt - m*costi) / (costt + m*costi) )**2
Rs = abs( (costi - m*costt) / (costi + m*costt) )**2

plot(ti,Rp,ti,Rs)
xlabel('Incident angle (degrees)')
ylabel('Reflectivity')
legend(('vertical pol','horizontal pol'),loc='best')
```

####If the incident radiation is unpolarized, then the parallel and perpendicular polarizations are present in equal amounts.

```python
Rtotal = 0.5 * (Rs + Rp)
plot(ti,Rtotal)
xlabel('Incident angle (degrees)')
ylabel('Total Reflectivity')
```

###Interesting features:

###1) The reflectivity of water is low for small incidence angles (say, 0 to 30 degrees) for most materials.  But it increases rapidly near 90 degrees.  THUS, flat surfaces of water are poor reflectors of solar radiation at high sun angles.  BUT they are excellent reflectors at sunrise and sunset.

###2) Horizontal polarization is typically much larger than vertical polarization.  THEREFORE, sunglasses are designed to block horizontal polarization and allow vertical polarization to pass through.

###3) The BREWSTER ANGLE is where the vertical polarization goes to 0.  At this angle, only horizontal polarization exists.

```python
# Brewster angle for water.
N1 = 1.000  # air
#N2 = 1.025  #Liquid He
N2 = 1.333  # water
#N2 = 1.544  # Sodium Chloride
#N2 = 2.419  # Diamond
m = N2/N1
tb = arcsin( ( m**2 / (m**2 + 1) )**0.5 )
print 'The Brewster angle is: ', tb*180./pi
```
