#Solar Spectral Irradiance

##Solar Spectral Irradiance at the top of the atmosphere and the surface

```python
import xlrd

# Reference Solar Spectral Irradiance: Air mass = 1.5
#   (http://rredc.nrel.gov/solar/spectra/am1.5/)
filename = '/Users/vonw/work/courses/WSU/2013-2014/spring/CE590/python/ASTMG173.xls'

wb = xlrd.open_workbook(filename)
sheets = wb.sheets()
wl   = array(sheets[0].col_values(0)[2:])
toa_flux = array(sheets[0].col_values(1)[2:])
sfc_flux = array(sheets[0].col_values(3)[2:])

figure()
plot(wl,toa_flux,wl,sfc_flux)
xlabel('Wavelength (nm)')
ylabel('Spectral Flux Density (W m-2 nm-1)')
title('NREL Reference Solar Irradiance\n (http://rredc.nrel.gov/solar/spectra/am1.5/)')
```

##Difference in the Solar Spectral Irradiance at TOA and the surface

###(Shows effect of absorption by atmospheric gases)

```python
figure()
plot(wl,toa_flux-sfc_flux)
xlabel('Wavelength (nm)')
ylabel('Flux Density Difference (W m-2 nm-1)')
title('Difference in TOA versus surface spectral flux density')
```
